# Basic Probability Terminology & Concepts

## <span style=color:skyblue>Experiment, Outcomes, and Sample Space</span>
### Experiment
An experiment is any action or process that generates outcomes.
**Example:** Observing Sachin’s performance in an ODI match is an experiment.

### Outcome
An outcome is a single result of an experiment.
**Example:** Sachin scoring 100 runs in a specific match is one possible outcome.

### Sample Space (S)
The sample space is the set of all possible outcomes of an experiment.
**Example:** For Sachin’s ODI career, the sample space might include all possible scores he could achieve in a match (0, 1, 2, ..., 200+).

## <span style=color:skyblue>Events</span>
An event is a subset of the sample space. It can contain one or more outcomes.
**Example:**
* Event A: Sachin scores more than 100 runs in a match.
* Event B: Sachin scores exactly 0 runs (a duck).

### Types of Events
1. Mutually Exclusive Events
Events that cannot happen at the same time.
Example: Sachin scoring a century and scoring a duck in the same match.
2. Joint Events
Events that can happen simultaneously.
Example: Sachin scores more than 50 runs, and the match is won by India.
3. Independent Events
Events where the occurrence of one does not affect the other.
Example: The outcome of Sachin’s performance and the weather conditions (assuming no match cancellations).
4. Exhaustive Events
A set of events that covers all possible outcomes in the sample space.
Example:
    * Event A: Sachin scores less than 50 runs.
    * Event B: Sachin scores 50 or more runs.
    * Together, these two events are exhaustive.

# Probability Rules

Understanding the fundamental rules of probability helps us compute the likelihood of events systematically. Let’s explore these rules using Sachin Tendulkar’s ODI career data.

## <span style=color:skyblue>Addition Rule</span>
The addition rule calculates the probability of either one event or another happening (the union of events).

### Formula:

`P(A∪B)=P(A)+P(B)−P(A∩B)`

**Example:**

What is the probability that Sachin scored more than 50 runs (Event A) or India won the match (Event B)?


````python
# What is the probability that Sachin Tendulkar scored more than 50 runs or India won the match in his ODI career?

# Total matches
total_matches = len(sachin_data)

# Event A: Sachin scored > 50 runs
event_A = sachin_data[sachin_data['runs'] > 50]

# Event B: India won the match
event_B = sachin_data[sachin_data['Won'] == True]

# Intersection (A ∩ B): Matches where Sachin scored > 50 runs and India won
intersection = pd.merge(event_A, event_B, how='inner')

# Probabilities
P_A = len(event_A) / total_matches
P_B = len(event_B) / total_matches
P_A_intersection_B = len(intersection) / total_matches

# Applying the addition rule
P_A_union_B = P_A + P_B - P_A_intersection_B

print(f"Probability of scoring > 50 runs or India winning: {P_A_union_B:.4f}")
````

### <span style=color:orange>Business Context</span>
A sports analytics company wants to quantify the likelihood of key match outcomes involving Sachin Tendulkar’s performance and team success. By calculating probabilities using the addition rule, they can:

* Determine how often Sachin scores above 50 or India wins, helping assess his impact on match outcomes.
* Inform predictive models for match results based on player performance.
* Provide statisticians, commentators, and fans with data-driven insights on performance scenarios.

## <span style=color:skyblue>Multiplication Rule</span>
The multiplication rule calculates the probability of both events happening (intersection of events).

### Formula:

`P(A∩B)=P(A)⋅P(B∣A)`

**Example:**

What is the probability that Sachin scored more than 50 runs (Event A) and India won the match (Event B)?

The probability of India winning depends on whether Sachin scores more than 50 runs.

We cannot directly calculate the joint probability 𝑃 ( 𝐴 ∩ 𝐵 ) without knowing 𝑃 ( 𝐵 ∣ 𝐴 )(the probability of India winning given Sachin's performance).

Without the formula, we'd have to experimentally determine how often India wins when Sachin scores more than 50, which is impractical with real-world data.


````python
## What is the probability that Sachin Tendulkar scored more than 50 runs and India won the match in his ODI career?
# Event A: Sachin scored > 50 runs
P_A = len(event_A) / total_matches

# Conditional probability: P(B | A)
conditional_data = sachin_data[sachin_data['runs'] > 50]
P_B_given_A = len(conditional_data[conditional_data['Won'] == True]) / len(event_A)

# Applying the multiplication rule
P_A_intersection_B = P_A * P_B_given_A

print(f"Probability of scoring > 50 runs and India winning: {P_A_intersection_B:.4f}")
````

### <span style=color:orange>Business Context</span>
A cricket analytics team wants to measure the likelihood of joint outcomes involving Sachin Tendulkar’s performance and team success. By applying the multiplication rule, they can:

Determine the probability that Sachin scores more than 50 and India wins, capturing the combined effect of player performance on match results.

Inform predictive modeling for match strategies, player impact analysis, and commentary insights.

Avoid impractical manual counting of joint occurrences across hundreds of matches by using conditional probabilities.


## <span style=color:skyblue>Marginal vs. Joint Probability</span>
* Marginal Probability: Probability of a single event happening (e.g., P(A) or P(B)).
Example: Probability that Sachin scored a century (Event C).
* Joint Probability: Probability of two events happening together (e.g., P(A ∩ B)).
Example: Probability that Sachin scored more than 50 runs and hit at least one six (Event D).

1. When We Can Directly Calculate Joint Probability. We can directly compute 𝑃 ( 𝐴 ∩ 𝐵 ) P(A∩B) when we have historical data or independent events.

Example: Rolling Two Dice (Independent Events) Let A = Getting a 4 on the first die.

Let B = Getting a 6 on the second die.

Since rolling one die does not affect the other, the events are independent:
```
𝑃 ( 𝐴 ∩ 𝐵 )
𝑃 ( 𝐴 ) × 𝑃 ( 𝐵 )
1 6 × 1 6
1 36
```
Here, we can directly multiply because the events are independent.

2. When We Cannot Directly Calculate Joint Probability. If events are dependent, we need extra information about how they influence each other.

Example: Sachin Scoring & India Winning (Dependent Events) If Sachin scores more than 50 runs (A), it increases the chance of India winning (B).

Here, we do not have 𝑃 ( 𝐴 ∩ 𝐵 ) directly.

Instead, we have:
```
𝑃 ( 𝐴 ∩ 𝐵 )
𝑃 ( 𝐴 ) × 𝑃 ( 𝐵 ∣ 𝐴 )
```
We need conditional probability because one event affects the other.


````python
#What is the marginal probability that Sachin Tendulkar scored a century, and what is the joint probability
# that he scored more than 50 runs and hit at least one six in an ODI match?

# Marginal probability: Sachin scored a century
event_C = sachin_data[sachin_data['century'] == True]
P_C = len(event_C) / total_matches

# Joint probability: Sachin scored > 50 runs and hit at least 1 six
event_D = sachin_data[(sachin_data['runs'] > 50) & (sachin_data['sixes'] > 0)]
P_D = len(event_D) / total_matches

print(f"Marginal Probability of Sachin scoring a century: {P_C:.4f}")
print(f"Joint Probability of scoring > 50 runs and hitting at least 1 six: {P_D:.4f}")
````

### <span style=color:orange>Business Context</span>
A cricket analytics firm wants to differentiate between single-event probabilities and combined-event probabilities in player performance:
* Marginal probability helps understand the likelihood of individual milestones, e.g., how often Sachin scores a century.
* Joint probability measures the chance of multiple achievements occurring together, e.g., scoring over 50 runs and hitting at least one six.

This analysis is crucial for:
* Performance modeling and predictive analytics.
* Evaluating player consistency and impact in high-stakes situations.
* Designing metrics for commentary, fan engagement, or fantasy cricket platforms.

## <span style=color:skyblue>Conditional Probability</span>

# Bayes' Theorem

Bayes' Theorem is a cornerstone of probability theory that helps us revise probabilities based on new evidence. It enables us to calculate conditional probabilities when the reverse probability and other relevant information are known.

## <span style=color:skyblue>Formula for Bayes' Theorem</span>
The formula for Bayes' Theorem is:
<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/167/013/original/Screenshot_2025-11-21_164529.png?1763723744" width="300" />

Where:

* P(A∣B): Probability of A occurring given that B has occurred (posterior).
* P(B∣A): Probability of B occurring given that A has occurred (likelihood).
* P(A): Probability of A occurring (prior).
* P(B): Probability of B occurring (evidence).

## <span style=color:skyblue>Intuition Behind Bayes' Theorem</span>
**Example:**
In Sachin’s dataset:
* Event A: Sachin scores a century.
* Event B: Sachin batted first.

We want to compute P(A∣B):

What is the probability that Sachin scores a century, given he batted first?


````python
## Given that Sachin Tendulkar batted first in an ODI match, what is the probability that he scored a century?

# Define the events
event_A = sachin_data[sachin_data['century'] == True]  # Event A: Scored a century
event_B = sachin_data[sachin_data['Inns'] == 1]        # Event B: Batted first

# Total matches
total_matches = len(sachin_data)

# Probabilities
P_A = len(event_A) / total_matches  # Prior probability of scoring a century
P_B = len(event_B) / total_matches  # Probability of batting first
P_B_given_A = len(event_A[event_A['Inns'] == 1]) / len(event_A)  # Likelihood

# Bayes' Theorem
P_A_given_B = (P_B_given_A * P_A) / P_B

print(f"Probability of scoring a century given Sachin batted first: {P_A_given_B:.4f}")
````

Bayes’ Theorem allows the team to combine historical performance data with contextual information for more accurate probability estimates.

- [ ] The probability that it was made by a skilled worker is 60%.
- [ ] The probability that it was made by an unskilled worker is 40%.
- [ ] Skilled and unskilled workers are equally likely, given the item is defective.

Explanation:


We need $P(\text{Unskilled} \mid \text{Defective})$.

* $P(\text{Defective} \mid \text{Skilled}) = 1 - 0.9 = 0.1$
* $P(\text{Defective} \mid \text{Unskilled}) = 1 - 0.6 = 0.4$

Now apply **Bayes’ Theorem**:

$$
P(\text{Skilled} \mid \text{Defective})=
\frac{P(\text{Skilled}) \cdot P(\text{Def} \mid \text{Skilled})}{P(\text{Def})}
$$

where

$$
P(\text{Def})=
(0.6 \cdot 0.1) + (0.4 \cdot 0.4)
=0.06 + 0.16
=0.22
$$

So:

$$
P(\text{Skilled} \mid \text{Def})=\frac{0.6 \cdot 0.1}{0.22}=
\frac{0.06}{0.22}
\approx 27.27%
$$

$$
P(\text{Unskilled} \mid \text{Def})
\frac{0.4 \cdot 0.4}{0.22}
\frac{0.16}{0.22}
\approx 72.73%
$$

Thus, **defective items are much more likely to come from unskilled workers → A is correct.**
